**Question 1 : Explain the difference between Absolute, Relative, and Mixed Cell
Referencing in Excel with examples ?**

> **1. Relative Cell Referencing:**

>> When we copy a formula that uses relative cell references, Excel automatically adjusts the cell references based on the new position of the formula. This is the defaylt behavior in Excel.

> **Example:**
>> If you have a formula =A1 + B1 in cell C1 and we copy it to cell C2, the formula in C2 will becomes =A2+B2. Excel shifts the references down one row to match the new position.

2. **Absolute Cell Referencing:**

>> absolute cell references remain constant, no matter where we copy or move the formula. This is useful when you want a formula to always refer to a specific cell. Use the dollar sign ($) before both the column letter and the row number like $A$1.

> **Example:**

>  If we have a formula =A1 + $B$1 in cell C1 and we copy it to cell C2, the formula in C2 will become =A2 + $B$1. The reference to B1 remains fixed, while A1 changes relatively.

> **3. Mixed Cell Referencing:**

>> Mixed cell references have either the column or the row fixed, but not both. This is helpful when we want to lock one part of the referance while allowing the other part to change.

> **Example:**

>> If w3e have a formula = $A1 * C1 in cell D1 and we copy it to cell D2, it becomes =$A2 * C2. The column A remains fixed, but the row number changes.

>** Example:**

> **Example:**

> if A1 contains "  Come Here ", we write =Trim(A1) then it will return "Come Here".

**Question 5 : Define the purpose of VLOOKUP and HLOOKUP. How are they different
from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best while usage?**

> Both VLOOKUP or (Vertical Lookup) and HLOOKUP or(Horizontal Lookup) are traditional Excel functions used to search for data in a table and return a corresponding value. They perform a 'lookup' operation. In short It's very use full function in excel for search values.

> **VLOOKUP:**

> Searches for a value in the first column of a table (vertically) and returns a value in the same row from a specified column. It's best used when our data is organized with lookup values in the leftmost column.

> **HLOOKUP: **

> Searches for a value in the first row of a table (horizontally) and returns a value in the same column from a specified row. It's used when our data is organised with lookup values in the top most row.

> **Comparision with XLOOKUP:**

> XLOOKUP is a newer, more flexible, and generally superior lookup function introduced in modern Excel versions. It addresses many of the limitations of VLOOKUP and HLOOKUP.
